```
total = 0
count = 0
ssim_value = 0

for x in a:
  for xx in b:
    def mse(imageA, imageB):
      # NOTE: the two images must have the same dimension
      err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
      err /= float(imageA.shape[0] * imageA.shape[1])
      return err

    def compare_images(imageA, imageB, title):
      # compute the mean squared error and structural similarity
      m = mse(imageA, imageB)
      s = measure.compare_ssim(imageA, imageB)
      return s

    original = cv2.imread('/content/drive/My Drive/Evaluation metrics/Fake_bad/'+x,0)
    Predicted = cv2.imread('/content/drive/My Drive/Evaluation metrics/Fake_bad/'+xx,0)

    ssim_value = compare_images(original, Predicted, "Original vs. Predicted")
    
    count = count + 1
    total = total + ssim_value
    print(ssim_value)    
    ssim_value = 0

    
print('Avg_SSIM:')
print(total/count)
```
